# res2next29_6cx24wx4scale的验证集loss

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df1 = pd.read_csv("./models/res2next29_6cx24wx4scale/log.csv")   #读取csv数据
print(df1)
```

```python
df2 = pd.read_csv('./models_isda/res2next29_6cx24wx4scale/log.csv')
print(df2)
```

```python
y1 = df1.iloc[:,5].tolist()       #获取dataFrame中的第3列，并将此转换为list
x = list(range(1,len(y1)+1))       #产生横轴坐标
y2 = df2.iloc[:,5].tolist()

# 创建图形并添加两条折线
fig, ax = plt.subplots()

# 白色折线
ax.plot(x, y1, color='black', label='res2net')

# 黑色折线
ax.plot(x, y2, color='red', label='res2net_isda')

# 添加图例
ax.legend()
plt.title("val loss comparsion")               #设置标题
plt.xlabel("epoch")                 #横轴名称
plt.ylabel("loss") 


plt.savefig('figs/res2next29_6cx24wx4scale_loss.svg',format='svg',dpi=300)
#fig.savefig(path+'输出图片.svg')#输出
#plt.show()
```
